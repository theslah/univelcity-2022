```python
import bs4
import requests
```

```python
soup = bs4.BeautifulSoup
```

```python
response  = requests.get("https://punchng.com/topics/news/")
html_data = soup(response.text) #parse html data into bs4
```

```python
news_cards = html_data.find_all("div") # find all div tags including the unwanted cards
news_cards
```

```python
for news_card in news_cards:
    print(news_card.text)
```

```python
news_cards = html_data.find_all("div", {"class", "grid-item"}) # find all div with specific attributes in this case with respect to class and grid item
news_cards
```

```python
for news_card in news_cards: # a more concise list for the articles
    print(news_card.text.strip())
```

```python
for news_card in news_cards: # a more concise list for links
    link_tag = news_card.find("a")
    print(link_tag.attrs["href"])
    print(news_card.text.strip())
```

```python
# go further to follow the links of each article and get the corresponding news from them

for news_card in news_cards: # a more concise list for links
    link_tag = news_card.find("a")
    link = link_tag.attrs["href"]
    
    link_response  = requests.get(link)
    link_html_data = soup(link_response.text)
    news = link_html_data.find("div", {"class":"entry-content"})
    
    print(link_tag.attrs["href"])
    print(news_card.text.strip())
    print(news.text.strip())
```

```python
# go further to follow the links of each article and get the corresponding news from them

url = "https://punchng.com/topics/news/page/{}/"

for page_number in range(1, 4):
    
    print(url.format(page_number))
    response = requests.get(url.format(page_number))
    html_data =soup(response.text) #parse html data into bs4
    
    news_cards = html_data.find_all("div", {"class":"grid-item"}) # find all div tags including the unwanted cards
    
    for news_card in news_cards: # a more concise list for links
        link_tag = news_card.find("a")
        link = link_tag.attrs["href"]

        link_response  = requests.get(link)
        link_html_data = soup(link_response.text)
        news = link_html_data.find("div", {"class":"entry-content"})

        print(link_tag.attrs["href"])
        print(news_card.text.strip())
        print(news.text.strip())
        
    
print(f"------------PAGE {page_number}------------")
      
# https://punchng.com/topics/news/page/2/
```

```python
# now write data into a folder with text files

url = "https://punchng.com/topics/news/page/{}/"

for page_number in range(1, 4):
    
    print(url.format(page_number))
    response = requests.get(url.format(page_number))
    html_data =soup(response.text) #parse html data into bs4
    
    news_cards = html_data.find_all("div", {"class":"grid-item"}) # find all div tags including the unwanted cards
    
    for news_card in news_cards: # a more concise list for links
        link_tag = news_card.find("a")
        link = link_tag.attrs["href"]

        link_response  = requests.get(link)
        link_html_data = soup(link_response.text)
        news = link_html_data.find("div", {"class":"entry-content"})

        print(link_tag.attrs["href"]) # link
        print(news_card.text.strip()) # title
        print(news.text.strip()) # article
       
        file_name = f"{news_card.text.strip()}"
       
        mode = "w"
        
        file = open(file_name, mode)

        file.write(link_tag.attrs["href"])
        
        file.write(news.text.strip())
    
print(f"------------PAGE {page_number}------------")
      
# https://punchng.com/topics/news/page/2/
```

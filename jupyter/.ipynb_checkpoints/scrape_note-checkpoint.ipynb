```python
import bs4
import requests
```

```python
soup = bs4.BeautifulSoup
```

```python
response = requests.get("https://hub360.com.ng/")
#response.text
```

```python
categories = soup(response.text).find_all("a")
categories
```

```python
categories[3]
```

```python
for category in categories:
    print(category.text)
```

```python
categories = soup(response.text).find_all("span")
categories
```

```python
for category in categories:
    print(category.text)
```
